## Exploratory Data Analysis (EDA)

### Explore data orders_df

Menggabungkan data products & data product_name_df

```python
product_name = pd.merge(
    left=products,
    right=product_name_df,
    how="inner",
    left_on="product_category_name",
    right_on="product_category_name"
)
product_name.head()
```

Menggabungkan data order_items & data product_name

```python
order_item_detail = pd.merge(
    left=order_items,
    right=product_name,
    how="inner",
    left_on="product_id",
    right_on="product_id"
)
order_item_detail.head()
```

Menggabungkan data orders & data order_item_detail

```python
orders_detail = pd.merge(
    left=orders,
    right=order_item_detail,
    how="inner",
    left_on="order_id",
    right_on="order_id"
)
orders_detail.head()
```

#### Explore data payments

Menggabungkan data orders & data payments

```python
payments_detail = pd.merge(
    left=orders,
    right=payments,
    how="inner",
    left_on="order_id",
    right_on="order_id"
)
payments_detail.head()
```

#### Explore data reviews

Menggabungkan data orders dan data reviews

```python
reviews_detail = pd.merge(
    left=orders,
    right=reviews,
    how="inner",
    left_on="order_id",
    right_on="order_id"
)
reviews_detail.head()
```

#### Explore All Data

Menggabungkan semua data

```python
data = pd.merge(
    left=orders_detail,
    right=payments_detail,
    how="inner",
    left_on=["order_id","customer_id","order_delivered_customer_date"],
    right_on=["order_id","customer_id","order_delivered_customer_date"]
)
data.head()
```

```python
all_data = pd.merge(
    left=data,
    right=reviews_detail,
    how="inner",
    left_on=["order_id","customer_id","order_delivered_customer_date"],
    right_on=["order_id","customer_id","order_delivered_customer_date"]
)
all_data.head()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Kategori produk apa yang paling sering dibeli dan paling jarang dibeli?

```python
all_data.head()
```

```python
sum_order_items = all_data.groupby("product_category_name_english").product_id.count().sort_values(ascending=False).reset_index()
sum_order_items.rename(columns={"product_id": "products"},inplace=True)
sum_order_items.head()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))

colors = ["#90CAF9", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

sn.barplot(
    x="products",
    y="product_category_name_english",
    data=sum_order_items.head(5),
    palette=colors,
    ax=ax[0]
)
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Produk yang paling sering dibeli", loc="center", fontsize=16)
ax[0].tick_params(axis ='y', labelsize=14)

sn.barplot(
    x="products",
    y="product_category_name_english", 
    data=sum_order_items.sort_values(by="products", ascending=True).head(5),
    palette=colors,
    ax=ax[1]
)
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Produk yang paling jarang dibeli", loc="center", fontsize=16)
ax[1].tick_params(axis='y', labelsize=14)

plt.suptitle("Kategori produk yang paling sering dibeli dan paling jarang dibeli", fontsize=20)
plt.show()
     
```

Dari grafik di atas, dapat dilihat bahwa kategori produk yang paling sering di beli adalah bed_bath_table, sedangkan produk yang paling jarang dibeli adalah security_and_services.

### Pertanyaan 2: Berapa banyak produk yang dibeli setiap bulannya?

```python
monthly_orders = all_data.resample(rule='M', on='order_delivered_customer_date').agg({"order_id": "size",})
monthly_orders
```

```python
monthly_orders.index = monthly_orders.index.strftime('%B')
monthly_orders = monthly_orders.reset_index()
monthly_orders.rename(columns={"order_id": "order_count",}, inplace=True)
monthly_orders
```

```python
monthly_orders = monthly_orders.sort_values('order_count').drop_duplicates('order_delivered_customer_date', keep='last')
monthly_orders
```

```python
month_map = {
    "January": 1,
    "February": 2,
    "March": 3,
    "April": 4,
    "May": 5,
    "June": 6,
    "July": 7,
    "August": 8,
    "September": 9,
    "October": 10,
    "November": 11,
    "December": 12
}

monthly_orders["month_num"] = monthly_orders["order_delivered_customer_date"].map(month_map)
monthly_orders = monthly_orders.sort_values("month_num")
monthly_orders = monthly_orders.drop("month_num", axis=1)
monthly_orders
```

```python
plt.figure(figsize=(10, 5))
plt.plot(
    monthly_orders["order_delivered_customer_date"],
    monthly_orders["order_count"],
    marker='o', 
    linewidth=2,
    color="#72BCD4"
)
plt.title("Banyak Pembelian Produk Per Bulan", loc="center", fontsize=20)
plt.xticks(fontsize=10, rotation=25)
plt.yticks(fontsize=10)
plt.show
```

Berdasarkan grafik di atas, terlihat bahwa metode pembayaran yang paling sering digunakan oleh pelanggan yaitu credit_card, diikuti dengan boleto, voucher, dan debit_card.
